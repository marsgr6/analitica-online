Una de las principales diferencias entre el enfoque de econometr铆a y la Ciencia de Datos por Machine Learning, es que por el enfoque de ML nos centramos en la anal铆tica predictiva m谩s que en las relaciones entre variables. Por ello, que a lo largo de este script no centraremos nuestros esfuerzos en interpretar coeficientes, o encontar relaciones entre variables, sino entrenar de manera 贸ptima a nuestros algoritmos para que alcancen niveles predictivos altos.



Como regla general, los algoritmos de ML son m谩s flexibles que el marco econom茅trico, ya que nos enfocamos en predecir, por ello muchos de los supuestos que cubrimos en la primera semana no aplican a este enfoque a excepci贸n de que si vamos a aplicar una Regresi贸n Lineal, requerimos haya linealidad en los par谩metros ya que scikit learn, que es el module en Python para modelar por ML, ajusta una l铆nea recta a los datos de entrenamiento.

### Conceptos

Dentro del enfoque de Machine Learning necesitaremos entender tres conceptos fundamentales:

#### Train-Test split

Es una t茅cnica que permite la divisi贸n de nuestra base de datos en dos categor铆as:

- La de de prueba llamada "test", que se utiliza para estimar el rendimiento de los algoritmos de aprendizaje autom谩tico  basados en predicciones.  De manera general, el conjunto de prueba se divide en el 30 % de los datos reales y constituye un conjunto de datos que el algoritmo nunca antes ha visto. De esta forma se garantiza que en un ambiente de producci贸n, el modelo desarrollado opere de manera similar a los resultados obtenidos en el "test" set
     
- La de de entrenamiento llamada "train", que se utiliza para que el algoritmo "aprenda" de las relaciones entre las variables y los datos, por lo que el conjunto de datos de entrenamiento es generalmente de mayor tama帽o en comparaci贸n con el conjunto de datos de prueba 


## M茅tricas de evaluaci贸n de la Regresi贸n Lineal

En lo que respecta a la evaluaci贸n de nuestro modelo, en el presente script ya no obtedremos el output donde evalu谩bamos los coeficientes y su significancia, si no que dependiendo del problema nos centraremos en m茅tricas que cuantifiquen el poder explicativo de nuestro modelo. As铆, si el problema es de regresi贸n, las tres medidas m谩s comunes utilizadas para evaluar las predicciones son:
- El Error cuadr谩tico medio (MSE)
- El Error cuadr谩tico medio (RMSE)
- El Error absoluto medio (MAE)

Las cuales de una u otra forma miden cuan lejos est谩n los valores predichos de los valores observados en nuestro "test set"

#### MSE

El error cuadr谩tico medio, o MSE  es una m茅trica de error, as铆 como una funci贸n de costos importante para algoritmos ajustados. Aqu铆, "m铆nimos cuadrados" se refiere a minimizar el error cuadr谩tico medio entre las predicciones y los valores esperados.

El MSE se calcula como la media o el promedio de las diferencias al cuadrado entre los valores objetivo predichos y esperados en un conjunto de datos y matem谩ticamente se expresa como:

$${MSE}=\frac{1}{n} \sum_{i=1}^n \left(Y_i-\hat{Y_i}\right)^2$$

Donde:
-  es el i-茅simo valor esperado en el conjunto de datos 
-  estimado, es el i-茅simo valor predicho

La diferencia entre estos dos valores se eleva al cuadrado, lo que tiene dos efectos:
- El efecto de eliminar el signo, lo que da como resultado un valor de error positivo, y 
- El efecto de inflar o magnificar errores grandes. Es decir, cuanto mayor sea la diferencia entre los valores pronosticados y esperados, mayor ser谩 el error cuadr谩tico positivo resultante, por lo que se penaliza m谩s a los modelos por errores m谩s grandes cuando MSE se usa como una funci贸n de costo


#### RMSE

El RMSE, es una extensi贸n del error cuadr谩tico medio que calcula la ra铆z cuadrada del error, lo que significa que las unidades del RMSE son las mismas que las unidades originales del valor objetivo que se predice, por ello si la variable objetivo tiene d贸lares como unidades, el RMSE tambi茅n tendr谩 la misma undiad y no "d贸lares al cuadrado" como el MSE.

El RMSE se define de la siguiente manera:

$$RMSE=\sqrt{\frac{1}{n} \sum_{i=1}^n \left(Y_i-\hat{Y_i}\right)^2}$$

#### MAE

El error absoluto medio, o MAE, es una m茅trica en que las unidades de la puntuaci贸n de error coinciden con las unidades del valor objetivo que se predice, a diferencia del RMSE los cambios en MAE son lineales. Por otra parte, y a diferencia de las m茅tricas previas, el MAE no da m谩s o menos peso a los diferentes tipos de errores, en cambio las puntuaciones del MAE aumentan linealmente con los incrementos en el error.

Como su nombre indica, la puntuaci贸n MAE se calcula como la media de los valores de error absolutos, por lo tanto, la diferencia entre un valor esperado y predicho puede ser positiva o negativa y se fuerza a ser positiva al calcular el MAE.

$$ MAE = \frac{1}{n} *{\sum_{i=1}^n} \left|Y_i-\hat{Y_i}\right|$$

## M茅tricas de evaluaci贸n de la Regresi贸n Log铆stica

Por su parte, si el problema es de clasificaci贸n, las m茅tricas m谩s comunes son la matriz de confusi贸n y cocientes derivados de esta. Entre los m谩s populares encontramos a:
- Accuracy
- Presici贸n
- Exhaustividad o "recall"
- F1 score

#### Accuracy

La precisi贸n en los problemas de clasificaci贸n es el n煤mero de predicciones correctas realizadas por el modelo sobre todos los tipos de predicciones realizadas.

$$ Accuracy = \frac{TP+TN}{TP+TN+FP+FN} $$

En el Numerador, est谩n nuestras predicciones correctas (Verdaderos positivos y Verdaderos negativos) y en el denominador, todas las predicciones hechas por el algoritmo (tanto las correctas como las incorrectas).

_驴Cu谩ndo usar el accuracy?_
El accuracy es es una buena medida cuando las clases de variables de destino en los datos est茅n casi equilibradas.

#### Precisi贸n o "precision"

La precisi贸n es una medida que nos da luces sobre la calidad del modelo ya que eval煤a, el n煤mero de elementos identificados correctamente como positivos sobre el total de positivos verdaderos.

$$ Precision = \frac{TP}{TP+FP} $$

#### Exhaustividad o "recall"

Es una medida que nos dice qu茅 cantidad el modelo de machine learning es capaz de identificar. Responde a la pregunta 驴qu茅 porcentaje de la categor铆a objetivo somos capaces de identificar?

$$ Recall = \frac{TP}{TP+FN} $$

#### F1 score

Es una puntuaci贸n 煤nica que represente tanto la Precisi贸n (P) como la exhaustividad o "recall".

$$ F1 = \frac{2*Precision*Recall}{Precision+Recall} = \frac{2*TP}{2*TP+FP+FN} $$

### 1. Inicializar el modelo

```python
from sklearn.linear_model import LinearRegression
```

### 2. Entrenar el modelo con los datos de entrenamiento

Antes de realizar nuestra divisi贸n, dejemos a nuestra base de datos en el estado deseado. Para ello, lo primero que vamos a hacer es separar nuestra variable objetivo de los predictores y recodificar aquellas variables categ贸ricas que no se encuentren dummificadas:

#### Recodificamos las variables categ贸ricas con Label Encoder

```python
var_cuantitativas = df.select_dtypes('number').columns
var_cualitativas  =df.select_dtypes('object').columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# creating instance of labelencoder
labelencoder = LabelEncoder()
```

```python
df[var_cualitativas] = df[var_cualitativas].apply(labelencoder.fit_transform)
```

Separamos a los predictores de nuestra variable objetivo:

```python
X = df[df.columns.difference(['precio'])]
y = df.precio
```

Como nuestra base de datos no contine la variable tiempo, podemos realizar un train/split aleatorio de la siguiente forma:

Importamos del module sklearn la funci贸n: train_test_split

```python
from sklearn.model_selection import train_test_split
```

Dividimos en dos partes a nuestra base de datos en un 80% para que el algoritmo entrene y un 20% para aplicar nuestras m茅tricas de desempe帽o.

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.20,random_state =123)
```

```python
print(X_train.shape,"",type(X_train))
print(y_train.shape,"\t ",type(y_train))
print(X_test.shape,"",type(X_test))
print(y_test.shape,"\t ",type(y_test))
```

# El Modelo de Regresi贸n Lineal por Sklearn

Una vez tenemos nuestros diferentes conjuntos de datos, procedemos a entrenar al modelo:

```python
modelo_regresion = LinearRegression()
modelo_regresion.fit(X_train, y_train)
```

### 3. Realizar las predicciones

```python
predicciones_train = modelo_regresion.predict(X_train)
predicciones_test = modelo_regresion.predict(X_test)
```

## M茅tricas de evaluaci贸n

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

#### MSE

```python
MSE_train = mean_squared_error(y_train, predicciones_train)
MSE_test = mean_squared_error(y_test, predicciones_test)
print(MSE_train)
print(MSE_test)
```

#### RMSE

```python
RMSE_train = np.sqrt(MSE_train)
RMSE_test = np.sqrt(MSE_test)
print(RMSE_train)
print(RMSE_test)
```

#### MAE

```python
MAE_train = mean_absolute_error(y_train, predicciones_train)
MAE_test = mean_absolute_error(y_test, predicciones_test)
print(MAE_train)
print(MAE_test)
```

#### R^2

```python
from sklearn.metrics import r2_score
```

```python
r_square_train = r2_score(y_train, predicciones_train)
r_square_test  = r2_score(y_test, predicciones_test)
print('El R^2 del subconjunto de entrenamiento es:' , r_square_train)
print('El R^2 del subconjunto de prueba es:' , r_square_test)
```

```python
# Print the Intercept:
print('intercepto:', modelo_regresion.intercept_)

# Print the Slope:
print('pendiente:', modelo_regresion.coef_) 
```

#### Predicciones es el subgrupo de datos de entrenamiento

#### Predicciones es el subgrupo de datos de prueba

```python
fig, ax = plt.subplots()
ax.plot(y_test.values)
ax.plot(predicciones_test)
plt.title("Valores observados vs. predichos en test set");
```

## Bonus: Feature importance

Uno de los aspectos importantes al momento de entregar un modelo, es dicernir las variabels conforme a su utilidad para predecir nuestra variable objetivo. Este tipo de an谩lisis se llama "feature importance" y calcula la importancia de cada variable independiente en nuestro modelo. Para ello, ajustamos nuestro modelo de regresi贸n lineal  y extraemos los coeficientes de los predictores, los cuales proporcionan la base para una puntuaci贸n de importancia de caracter铆sticas.

Nota: Para este tipo de an谩lisis requerimos que las las variables de entrada tengan la misma escala o se hayan escalado antes.

Requerimos normalizar las variables explicativas porque en un an谩lisis de feature importance si lo obviamos, el modelo dif铆cilmente captar谩 la contribuci贸n de las variables de menor escala, incluso si son fuertes. Para ello, usaremos la funci贸n StandardScaler del module de sklearn.

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
```

Observan algo interesante en el c贸digo precedente?

### Entrenar el modelo con los datos de entrenamiento

Una vez tenemos nuestros diferentes conjuntos de datos, procedemos a entrenar al modelo:

```python
modelo_regresion_std = LinearRegression()
modelo_regresion_std.fit(X_train_std, y_train)
```

### Realizar las predicciones

```python
predicciones_train_std = modelo_regresion_std.predict(X_train_std)
predicciones_test_std = modelo_regresion_std.predict(X_test_std)
```

#### R^2

```python
r_square_train_std = r2_score(y_train, predicciones_train_std)
r_square_test_std  = r2_score(y_test, predicciones_test_std)
print('El R^2 del subconjunto de entrenamiento es:' , r_square_train_std)
print('El R^2 del subconjunto de prueba es:' , r_square_test_std)
```

#### Predicciones es el subgrupo de datos de entrenamiento

```python
fig, ax = plt.subplots()
ax.plot(y_train.values)
ax.plot(predicciones_train_std)
plt.title("Valores observados vs. predichos en train set");
```

#### Predicciones es el subgrupo de datos de prueba

```python
fig, ax = plt.subplots()
ax.plot(y_test.values)
ax.plot(predicciones_test_std)
plt.title("Valores observados vs. predichos en test set");
```

```python
importancia = modelo_regresion_std.coef_
```

```python
# Resumen
for i,v in enumerate(importancia):
    print('Variable explicativa No. %0d, Score: %.5f' % (i,v))
```

```python
# Graficar la importancia o "feature importance"
plt.bar([x for x in range(len(importancia))], importancia)
plt.show()
```

Un valor de "feature importance" negativo significa que esa caracter铆stica hace que aumente la funci贸n de costos, por lo que el modelo no est谩 hace un buen uso de esa caracter铆stica.

# El Modelo de Regresi贸n Lo铆stica por Sklearn

Una vez tenemos nuestros diferentes conjuntos de datos, procedemos a re-extraer nuestras variables explicativas y variable objetivo.

```python
X_logit = df[df.columns.difference(['AC'])]
y_logit = df.AC
```

```python
print(X_train_logit.shape,"",type(X_train_logit))
print(X_train_logit.shape,"\t ",type(y_train_logit))
print(X_test_logit.shape,"",type(X_test_logit))
print(X_test_logit.shape,"\t ",type(y_test_logit))
```

### Entrenar el modelo con los datos de entrenamiento

### M茅tricas de evaluaci贸n

#### Matriz de confusi贸n

```python
from sklearn import metrics
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
cm = metrics.confusion_matrix(y_test_logit, predicciones_test_logit)
print(cm)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.matshow(cm, cmap=plt.cm.Oranges, alpha=0.3)
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(x=j, y=i,s=cm[i, j], va='center', ha='center', size='xx-large')
 
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```

#### Precisi贸n o "precision"

```python
print('Precision: %.3f' % precision_score(y_test_logit, predicciones_test_logit))
```

La precisi贸n es una medida que nos da luces sobre la calidad del modelo ya que eval煤a, el n煤mero de elementos identificados correctamente como positivos sobre el total de positivos verdaderos.

#### Exhaustividad o "recall"

```python
print('Recall: %.3f' % recall_score(y_test_logit, predicciones_test_logit))
```

#### F1 score

```python
print('F1 Score: %.3f' % f1_score(y_test_logit, predicciones_test_logit))
```
